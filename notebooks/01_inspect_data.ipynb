```python
# ==========================================
# SCRIPT CLEANING V2.0 (Updated for FIFA 24)
# ==========================================

# 1. Buang Kiper (GK)
# Di dataset ini nama kolomnya 'player_positions'
# Kita buang jika di dalamnya mengandung kata 'GK'
df = df[~df['player_positions'].str.contains('GK', na=False)].copy()

# 2. Fitur Seleksi (Disesuaikan dengan nama kolom baru kamu)
# Metadata (Info Pemain)
meta_cols = ['player_id', 'short_name', 'long_name', 'player_positions', 
             'overall', 'potential', 'age', 'club_name', 'league_name', 
             'nationality_name', 'value_eur', 'wage_eur', 'player_url'] # Ganti real_face jadi url buat foto nanti

# Skill Teknis (Angka untuk AI)
# Perhatikan: nama kolomnya lebih panjang di dataset ini
skill_cols = [
    'pace', 'shooting', 'passing', 'dribbling', 'defending', 'physic',
    'attacking_crossing', 'attacking_finishing', 'attacking_heading_accuracy',
    'attacking_short_passing', 'attacking_volleys',
    'skill_dribbling', 'skill_curve', 'skill_fk_accuracy', 'skill_long_passing',
    'skill_ball_control',
    'movement_acceleration', 'movement_sprint_speed', 'movement_agility',
    'movement_reactions', 'movement_balance',
    'power_shot_power', 'power_jumping', 'power_stamina', 'power_strength',
    'power_long_shots',
    'mentality_aggression', 'mentality_interceptions', 'mentality_positioning',
    'mentality_vision', 'mentality_penalties', 'mentality_composure',
    'defending_marking_awareness',
    'defending_standing_tackle', 'defending_sliding_tackle'
]

# Gabungkan
final_cols = meta_cols + skill_cols
df_clean = df[final_cols].copy()

# 3. Hapus Baris yang Skill-nya Kosong (Null)
df_clean = df_clean.dropna(subset=skill_cols)

# 4. Reset Index
df_clean.reset_index(drop=True, inplace=True)

print(f"Data Bersih Siap Pakai: {df_clean.shape}")
display(df_clean.head())
```
